```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../data/asthma_disease_data.csv')

print("Dataset Shape:", df.shape)
display(df.head())

print("\nData Types and Non-Null Counts:")
print(df.info())

print("\nSummary Statistics:")
display(df.describe())

print("Missing Values per Column:")
print(df.isnull().sum())
```

```python
print("Number of Duplicate Rows:", df.duplicated().sum())
```

```python
categorical_cols = ['Gender', 'Ethnicity', 'EducationLevel', 'Smoking', 'PetAllergy', 
                    'FamilyHistoryAsthma', 'HistoryOfAllergies', 'Eczema', 'HayFever', 
                    'GastroesophagealReflux', 'Wheezing', 'ShortnessOfBreath', 
                    'ChestTightness', 'Coughing', 'NighttimeSymptoms', 
                    'ExerciseInduced', 'Diagnosis', 'DoctorInCharge']

for col in categorical_cols:
    print(f"\n{col} Value Counts:")
    print(df[col].value_counts())
```

```python
numerical_cols = ['Age', 'BMI', 'PhysicalActivity', 'DietQuality', 'SleepQuality', 
                  'PollutionExposure', 'PollenExposure', 'DustExposure', 
                  'LungFunctionFEV1', 'LungFunctionFVC']

df[numerical_cols].hist(bins=20, figsize=(15, 10))
plt.tight_layout()
plt.show()
```

```python
# Outliers check
plt.figure(figsize=(12, 6))
sns.boxplot(data=df[numerical_cols])
plt.xticks(rotation=45)
plt.title("Boxplot of Numerical Variables")
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.heatmap(df[numerical_cols + ['Diagnosis']].corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title("Correlation Matrix")
plt.show()
```

```python
sns.countplot(x='Diagnosis', data=df)
plt.title("Distribution of Diagnosis (0: No Asthma, 1: Asthma)")
plt.show()
```

```python
for col in numerical_cols:
    plt.figure()
    sns.boxplot(x='Diagnosis', y=col, data=df)
    plt.title(f"{col} by Diagnosis")
    plt.show()
```

```python
# Crosstab
for col in categorical_cols:
    if col != 'Diagnosis':
        ct = pd.crosstab(df[col], df['Diagnosis'])
        print(f"\n{col} vs Diagnosis:")
        print(ct)
```
